```python
import pandas as pd

data_filepath = "data-unclean.csv"
```

```python
data = pd.read_csv(data_filepath)

pd.set_option("display.max_columns", None)
data.head()
```

# Fixing DtypeWarning: Columns (15) have mixed types

```python
data = pd.read_csv(data_filepath, dtype={"address": str})

print(f"COUNT OF TYPES BEFORE:\n{data["address"].apply(type).value_counts()}\n------------------------")

problematic_rows = data[~data["address"].apply(lambda address: isinstance(address, str))]
data["address"] = data["address"].fillna("Unknown") # CLEANING

print(f"COUNT OF TYPES AFTER:\n{data["address"].apply(type).value_counts()}")
```

# Dropping columns

```python
# dropping
data = data.drop(["title", "body", "amenities", "has_photo", "price_display", "source"], axis=1)

date = pd.to_datetime(data["time"], unit="s").dt.strftime("%m/%d/%Y")
time = pd.to_datetime(data["time"], unit="s").dt.strftime("%H:%M %p")
data = data.drop("time", axis=1)
data["date"] = date
data["time"] = time

data.head()
```

# Renaming columns

```python
# renaming
data = data.rename(columns={
    "pets_allowed": "pets allowed",
    "price_type": "price type",
    "square_feet": "square feet",
    "cityname": "city"
})

data.head()
```

# Changing NaN values

```python
data.info()
```

```python
data = data.fillna({
    "bathrooms": "Unknown",
    "bedrooms": "Unknown",
    "pets allowed": "Unknown",
    "price": "Unknown",
    "city": "Unknown",
    "state": "Unknown",
    "latitude": "Unknown",
    "longitude": "Unknown",
})
```
